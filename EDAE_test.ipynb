# Run EDAE 
- To evaluate the performance of EDAE, we provide a demo file here.
- The full paper can be found in Pattern Recognition Letter or arXiv.
- The test was conducted based on the environment of ADBench, please refer to https://github.com/Minqi824/ADBench for details.

```python
# import basic package
import os
import pandas as pd
import re
import warnings
warnings.filterwarnings("ignore")

from data_generator import DataGenerator
from myutils import Utils

datagenerator = DataGenerator() # data generator
utils = Utils() # utils function

dataset_EDAE = ['1_abalone', '3_annthyroid', '7_Cardiotocography', '10_cover', '14_Hepatitis', 
                '18_Ionosphere', '19_landsat', '22_magic.gamma', '44_WDBC', '45_Wilt' ]
```

```python
from baseline.Unsupervised_autoencoder.run import AutoEncoder
from baseline.EDAE.run import EDAE
from baseline.DevNet.run import DevNet
from baseline.FEAWAD.run import FEAWAD

model_dict = {"EDAE": EDAE, "DevNet": DevNet, "FEAWAD": FEAWAD, "AutoEncoder": AutoEncoder}

df_AUCROC = pd.DataFrame(data=None, index=dataset_EDAE, columns = model_dict.keys())
df_AUCPR = pd.DataFrame(data=None, index=dataset_EDAE, columns = model_dict.keys())
```

```python
# seed for reproducible results
seed = 42

for dataset in dataset_EDAE:
    
    #la: ratio of labeled anomalies, from 0.0 to 1.0
    #realistic_synthetic_mode: types of synthetic anomalies, can be local, global, dependency or cluster
    #noise_type: inject data noises for testing model robustness, can be duplicated_anomalies, irrelevant_features or label_contamination
    
    # import the dataset
    datagenerator.dataset = dataset # specify the dataset name
    data = datagenerator.generator(la=0.05, realistic_synthetic_mode=None, noise_type=None) # only 5% labeled anomalies are available
    
    for name, clf in model_dict.items():   
        clf = clf(seed=seed, model_name=name)
        
        # training
        clf = clf.fit(X_train=data['X_train'], y_train=data['y_train'])
        
        # output predicted anomaly score on testing set
        score = clf.predict_score(data['X_test'])
        
        # evaluation
        result = utils.metric(y_true=data['y_test'], y_score=score)

        # save results
        df_AUCROC.loc[dataset, name] = result['aucroc']
        df_AUCPR.loc[dataset, name] = result['aucpr']
```
